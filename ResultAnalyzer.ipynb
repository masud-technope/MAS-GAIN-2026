```python
import os, io
import numpy as np
import pandas as pd
```

```python
def get_comparision_stats(lines):
    votes=dict()
    for line in lines:
        parts=line.strip().split(',')
        if len(parts) == 5:
            label=parts[1].strip()
            if label in votes:
                votes[label] += 1
            else:
                votes[label] = 1
    return votes
```

```python
def get_result_files(repo_name):
    delegate_fault_file=f"dataset/result-compare/{repo_name}-delegate.txt"
    message_lost_file=f"dataset/result-compare/{repo_name}-message-lost.txt"
    message_corrupt_file=f"dataset/result-compare/{repo_name}-message-corrupt.txt"
    return delegate_fault_file, message_lost_file, message_corrupt_file
```

```python
def get_file_content(file_path):
    with open(file_path, "r") as f:
        return f.readlines()
```

```python
def show_fault_details(repo_name, fault_file,):
    votes = get_comparision_stats(get_file_content(fault_file))
    print(f"Repository: {repo_name}")
    print(f"Votes: {votes}")
```

```python
delegate_fault_file, message_lost_file, message_corrupt_file = get_result_files("fastapi")
show_fault_details("fastapi", delegate_fault_file)
show_fault_details("fastapi", message_lost_file)
show_fault_details("fastapi", message_corrupt_file)
```

```python
delegate_fault_file, message_lost_file, message_corrupt_file = get_result_files("pandas")
show_fault_details("pandas", delegate_fault_file)
show_fault_details("pandas", message_lost_file)
show_fault_details("pandas", message_corrupt_file)
```
